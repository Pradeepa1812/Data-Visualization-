# Import Seaborn Library

```python
import seaborn as sns
```

```python
%matplotlib inline
```

# Load DataSet

```python
iris=sns.load_dataset('iris')
```

```python
iris.head()
```

```python
iris.columns
```

```python
iris.shape
```

```python
iris.describe()
```

# Distplot

```python
sns.distplot(iris['sepal_length'])
```

```python
sns.distplot(iris['sepal_length'],kde=False,bins=30)
```

# JointPlot

```python
sns.jointplot(x='sepal_length',y='sepal_width',data=iris,kind='hex')
```

```python
sns.jointplot(x='petal_length',y='sepal_width',data=iris,kind='reg')
```

```python
sns.jointplot(x='sepal_length',y='sepal_width',data=iris,kind='kde')
```

# Pairplot

```python
sns.pairplot(iris,hue='species',palette='bright')
```

# Rugplot

```python
sns.rugplot(iris['petal_length'])
```

# Kde Plot

```python
sns.kdeplot(iris['sepal_width'],lw=5)
```

# Categorical data

# BarPlot

```python
import numpy as np
```

```python
sns.barplot(x='species',
          y='sepal_width',
          data=iris,
           estimator=np.std)
```

# CountPlot

```python
sns.countplot(x='species',
             data=iris)
```

# BoxPlot

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(5,5))
sns.boxplot(x='species',y='petal_length',data=iris)
```

```python
plt.figure(figsize=(20,5))
sns.boxplot(x='petal_width',y='petal_length',data=iris,hue='species')
```

# Violin Plot

```python
sns.violinplot(x='species',y='petal_length',data=iris)
```

```python
plt.figure(figsize=(15,5))
sns.violinplot(x='petal_width',y='petal_length',data=iris,hue='species')
plt.legend(loc=2)
```

# Stripplot

```python
sns.stripplot(x='species',y='sepal_width',data=iris,jitter=True)
```

# swarm Plot

```python
sns.swarmplot(x='species',y='petal_length',data=iris)
```

```python
sns.violinplot(x='species',y='petal_length',data=iris)
sns.swarmplot(x='species',y='petal_length',data=iris,color='purple')
```

# Grid

```python
import pandas as pd
```

```python
exercise=sns.load_dataset('exercise',index_col=0)
```

```python
exercise.head()
```

```python
exercise['kind'].unique()
```

```python
g=sns.PairGrid(exercise)
g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
plt.grid(g)
g.map_lower(sns.kdeplot)
plt.grid(g)
```

# FacetGrid

```python
g=sns.FacetGrid(data=exercise,col='time',row='diet')
g.map(sns.distplot,'pulse')
```

# Regression Plot

```python
sns.lmplot(data=exercise,x='pulse',y='id',hue='diet',markers=['o','^'],aspect=1,size=6)
```
