# Libraries

```python
from pandas_datareader import data,wb
import pandas as pd
import numpy as np
import datetime
%matplotlib inline
```

# Using datareader we read the data from online

```python
start=datetime.datetime(2006,1,1)
end=datetime.datetime(2016,1,1)
BAC=data.get_data_yahoo("BAC",start,end)
GS=data.get_data_yahoo("GS",start,end)
C=data.get_data_yahoo("C",start,end)
JPM=data.get_data_yahoo("JPM",start,end)
MS=data.get_data_yahoo("MS",start,end)
WFC=data.get_data_yahoo("WFC",start,end)
```

```python
C
```

# Create a list of the tickers in alphabetical order

```python
tickers=['BAC','C','GS','JPM','MS','WFC']
```

# Concat

```python
bank_stocks=pd.concat([BAC,C,GS,JPM,MS,WFC],axis=1,keys=tickers)
```

```python
bank_stocks.head()
```

# setting the column name

```python
bank_stocks.columns.names=['Bank Tickers','Stock Info']
```

```python
bank_stocks.head()
```

# EDA

# Max  Open price for each bank_stock throughout the time period

```python
#for tick in tickers:
#  print(tick,bank_stocks[tick]['Open'].max())
    
bank_stocks.xs(key='Open',axis=1,level='Stock Info').max()
```

# creating new empty dataframe called returns

```python
returns=pd.DataFrame()
```

# we can use pandas pct_change() method  on the Open column to create a column representing this return value

```python
for tick in tickers:
    returns[tick+'Return']=bank_stocks[tick]['Close'].pct_change()
```
